```python
import pandas as pd 
data = {
    "Time":[1,2,3,4,5],
    "Value":[10,None,20,None,30]
}
# use interpolate() method to fill None data through analysing the trend 
df = pd.DataFrame(data)
print("Before interpolation: ")
print(df)
df['Value'] = df['Value'].interpolate(method="linear")
print("After interpolation: ")
print(df)


# 1- timer series data
# 2- numeric data with trend
# 3-  void dropping rows

```
