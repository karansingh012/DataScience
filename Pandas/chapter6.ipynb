# Missing.py

```python
import pandas as pd 
data = {
    "Name": ['Ram', None, 'Ghanshyam', 'Dhanshyam','Aditi','Jagdish', 'Raj', 'Simran'],
    "Age": [28,None,22,30, 29,40,25,32],
    "Salary": [50000,None,45000,52000,49000, 70000,48000,58000],
    "Performance_Score": [85,90,78,92, 88,95,80,89]
}
df = pd.DataFrame(data)
print(df)
print(df.isnull().sum())
```

# Handle.py

```python
# missing ko values ko hatane ke liye 
# df.dropna(inplace=True) yha per hum axis ki value dal kr row wise and column wise kr skte hai 
import pandas as pd 
data = {
    "Name": ['Ram', None, 'Ghanshyam', 'Dhanshyam','Aditi','Jagdish', 'Raj', 'Simran'],
    "Age": [28,None,22,30, 29,40,25,32],
    "Salary": [50000,None,45000,52000,49000, 70000,48000,58000],
    "Performance_Score": [85,90,78,92, 88,95,80,89]
}
df = pd.DataFrame(data)
print(df)
df.dropna(inplace=True)
print(df)
```

```python
#fill Nan value 
import pandas as pd 
data = {
    "Name": ['Ram', None, 'Ghanshyam', 'Dhanshyam','Aditi','Jagdish', 'Raj', 'Simran'],
    "Age": [28,None,22,30, 29,40,25,32],
    "Salary": [50000,None,45000,52000,49000, 70000,48000,58000],
    "Performance_Score": [85,90,78,92, 88,95,80,89]
}
df = pd.DataFrame(data)
print(df)
# df.fillna(0,inplace=True)
# print(df)
# agr nan vlaue ki jgh hume mean fill krna hai like agr salary etc
df['Age'].fillna(df['Age'].mean(),inplace=True)
print(df)
# same aaise hum salary me bhi kr skte hai 
df['Salary'].fillna(df['Salary'].mean(),inplace=True)
print(df)
```
