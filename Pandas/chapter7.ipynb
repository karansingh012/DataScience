# Shorting & Aggregation

```python
# sorting data
#sorting data one column sort_values()
#df.sort_values(by="column name",True/False, inplace=True)
import pandas as pd 
Data = {
    "Name":['karan','mayank','prabhat','faizan'],
    "Age":[19,22,23,20],
    "Salary":[80000,50000,70000,60000]
}
df = pd.DataFrame(Data)
df.sort_values(by="Salary",ascending=True,inplace=True)
print("Sorted salary by Descending: ")
print(df)
```

```python
## multiple column ko sort krna 
import pandas as pd 
Data = {
    "Name":['karan','mayank','prabhat','faizan'],
    "Age":[19,22,23,20],
    "Salary":[80000,50000,70000,60000]
}
df = pd.DataFrame(Data)
df.sort_values(by=["Age","Salary"],ascending=[True,False],inplace=True)
print("Sorted salary by Descending: ")
print(df)
```

```python
# Aggregation means summary 
# df["column_name"].mean()
# df["column_name"].sum()
# df["column_name"].min()
# df["column_name"].max()
import pandas as pd 
Data = {
    "Name":['karan','mayank','prabhat','faizan'],
    "Age":[19,22,23,20],
    "Salary":[80000,50000,70000,60000]
}
df = pd.DataFrame(Data)
avg_salary = df['Salary'].mean()
print(avg_salary)
```
