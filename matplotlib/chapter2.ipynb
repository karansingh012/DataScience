```python
#plt.subplot(nrows,ncols,index)
import matplotlib.pyplot as plt 
x = [1,2,3,4]
y = [10,20,5,25]
plt.subplot(1,2,1) #1 row , 2 column, 1st subplot
plt.plot(x,y)
plt.title("linear chart")


plt.subplot(1,2,2) # 1row , 2nd column , 2nd subplot
plt.bar(x,y)
plt.title("Bar chart")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt 
#fig, ax = plt.subplot(nrows,ncols,figsize=(width,height))
fig, ax = plt.subplots(1,2,figsize = (10,5))
x = [1,2,3,4]
y = [10,20,5,25]
ax[0].plot(x,y)
ax[0].set_title("line plot")

ax[1].bar(x,y)
ax[1].set_title("bar chart")
plt.tight_layout() ## pure screen me aa jayega 

plt.show()
```
