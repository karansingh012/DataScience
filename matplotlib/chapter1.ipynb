```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [10,15,30,40,50]
plt.plot(x,y)
plt.show
```

```python
import matplotlib.pyplot as plt
x = ['mon','tues','wed','thurs','frid','satur']
y = [8,10,12,14,18,22]
plt.plot(x,y)
plt.title("Bakery sales this week")
plt.xlabel("Day of the week")
plt.ylabel("sales per day")
plt.grid() # for shwoing grid
plt.show()
```

```python
import matplotlib.pyplot as plt
months = [1,2,3,4]
sales = [1000,1300,2000,800]
plt.plot(months,sales,color = 'skyblue',linestyle = "--",linewidth = 2, marker = 'o',label = "Sales data 2025" )
plt.title("sales analysis")
plt.xlabel("months")
plt.ylabel("sales per months")
plt.legend(loc='upper left',fontsize = 12)
plt.grid(color='gray',linestyle=':',linewidth=1)
plt.xlim(1,4)
plt.ylim(0,2000)
plt.xticks([1,2,3,4],['m1','m2','m3','m4'])
plt.show()
```

```python
products = ['A','B','C','D']
sales = [1200,3000,2400,800]
plt.bar(products,sales,color="orange",label="Sales data 2025")
plt.title("products sales comparison")
plt.xlabel("products")
plt.ylabel("sales")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
region = ['East','West','South','North']
revenue = [1000,1500,1800,2300]
plt.pie(revenue,labels=region,autopct='%1.1f%%',colors=['gold','skyblue','lightgreen','red'])
plt.title("revenue contribution by region")
plt.show()
```

```python
import matplotlib.pyplot as plt
scores = [53,45,23,45,89,67,33,17,98,67,56,45,54,34,43]
plt.hist(scores,bins=5,color = "purple",edgecolor='black')
plt.title("scores distribution of student")
plt.xlabel("scores range")
plt.ylabel("student")
plt.show()
```

```python
import matplotlib.pyplot as plt
hours_studies = [2,3,4,5,6,7,8]
exam_scores = [50,55,60,65,70,75,85]
plt.scatter(hours_studies,exam_scores,color='green',marker='o',label="student data")

plt.xlabel('hours studies')
plt.ylabel('exam score')
plt.title("Relationship between study time exam score")
plt.legend()
plt.grid()
plt.show()
```
