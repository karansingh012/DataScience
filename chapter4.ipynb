# Advanced Numpy
# 1. Insert

```python
"""
np.insert(array, index , value, asix=None)
array - original array
index -
Value -
axis -
axis = 0, row wise
axis = 1, column wise 
"""
import numpy as np
arr = np.array([10,20,30,40,50,60])
new_arr = np.insert(arr,2,100,axis=0) # 30 ko change krke ussme 100 dlna hai hume theek 
print(new_arr)
```

```python
print("Before Inserting")
arr_2d = np.array([[10,20],[30,40]])
print(arr_2d)
print("After inserting")
new_arr = np.insert(arr_2d,1,[5,6],axis = 1)
print(new_arr) # agr axis ki value zero kr doge toh ye array ke flaten version me fill kr dega 
```

# Append.py

```python
arr = np.array([20,30,40,50])
new_arr = np.append(arr,[40,50,60])
print(new_arr)
```

# concatenate((array1,array2),axis = 0)

```python
"""
axis 0 > vertical stacking
axis 1 > horizontal stacking
"""
arr1 = np.array([20,30,40])
arr2 = np.array([20,30,40])
new_array = np.concatenate((arr1,arr2))
print(new_array)
```

# np.delete.py


```python
arr = np.array([10,20,30,40])
new_array =np.delete(arr,1)
print(new_array)
```

```python
arr_2d = np.array([[10,20],[30,40]])
new_array_2d = np.delete(arr_2d,1,axis = 0)
print(new_array_2d)
```

# Stacking.py

```python
"""
vstack() row wise
hstack() column wise
"""
arr1 = np.array([20,30,40])
arr2 = np.array([20,30,40])
print(np.vstack((arr1,arr2)))
print(np.hstack((arr1,arr2)))
```
