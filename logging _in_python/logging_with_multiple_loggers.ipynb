* You can create multiples logger for different parts of your applications

```python
import logging
## Create a logger for module1
logger1 = logging.getLogger("module1")
logger1.setLevel(logging.DEBUG)

## Create a logger for module2
logger2 = logging.getLogger("module2")
logger2.setLevel(logging.WARNING)

# Now configure logging fresh
logging.basicConfig(
    level=logging.DEBUG,
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',
    datefmt='%Y-%m-%d %H:%M:%S'
)
```

```python
## log message with different loggers 
logger1.debug("This is my debug message for module1")
logger2.warning("This is my warning message for module2")

```
