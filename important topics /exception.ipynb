```python
# use of exception handling in python : why we use it?
# because sometimes our code may give error due to various reasons like wrong input from user, file not found , division by zero etc. so to avoid such errors we use exception handling.
# Exception try, except block
try:
    a=b
except:
    print("the variable has not been assigned")
```

```python
try:
    num = int(input("Enter a number"))
    result = 10/num
except ValueError:
    print("This is not a valid number")
except ZeroDivisionError:
    print("enter the denominator grater than 0")
except Exception as ex:
    print(ex)
```

```python
## try, except,else block
try:
    num = int(input("Enter a number"))
    result = 10/num
except ValueError:
    print("this is not a valid number")
except ZeroDivisionError:
    print("You can't divide by zero!")
except Exception as ex:
    print(ex)
else:
    print(f"the result is {result}")
```

```python
## try, except,else block and finally
try:
    num = int(input("ENter a number: "))
    rsult = 10/num
except ValueError:
    print("That's not a valid number!")
except ZeroDivisionError:
    print("You can't divide by zero!")
except Exception as ex:
    print(ex)
else:
    print(f"the result is {result}")
finally:
    print("Execution complete")
```

```python
# file handling and exception handling 
try:
    file = open('example1.txt','r')
    content = file.read()
    print(content)
except FileNotFoundError:
    print("The file does not exists")
finally:
    if 'file' in locals() and not file.closed():
        file.close()
        print("file close")
```
