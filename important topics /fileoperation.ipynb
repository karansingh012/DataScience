```python
## work of with open statement is to open a file and automatically close it after the nested block of code is executed.
## why we use it : to avoid memory leaks and to ensure that file is properly closed after its suite finishes, even if an exception is raised on the way.
# file: read_example.py
with open('example.txt','r') as file:
    content = file.read()
    print(content)
```

```python
# another way to read file line by line
with open('example.txt','r') as file:
    for line in file: ## forloop will read line by line
        # print(line)
        print(line.strip()) # this strip() function is remove new line character 
```

```python
# write a file(overwrite)
with open('example.txt','w') as file:
      file.write("Hello world \n")
      file.write("karan is my name\n")
```

```python
# write a file(without overwrite)
with open('example.txt','a') as file:
    file.write("append operation taking place!\n")
```

```python
# how to write multiple line 
lines = ['first line\n','second line \n','third line \n']
with open('example.txt','a') as file:
    file.writelines(lines)
```

```python
# how to work binary file 
# write to binaary file 
data = b'01010110 11100011 00101101 11001010 00011101 10110100'
with open('example.bin','wb') as file:
    file.write(data)
```

```python
with open("example.bin",'br') as file:
    content = file.read()
    print(content)
```

```python
# i want to read the content from a source text file and write a destination text file 
# copy a text file 
with open('example.txt','r') as file:
    content = file.read()
with open('destination.txt','w') as destination_file:
    destination_file.write(content)
```

```python
# writing and then reading file 
with open("example.txt",'w+') as file:
    file.write("HEllo world\n")
    file.write("THis is a new line\n")

    # move the file cursor to the bigning 
    file.seek(0)
    # read the content of the file 
    content = file.read()
    print(content)
```

```python
## R+ mode : it is used to read and write the file without truncating the file
with open('example.txt','r+') as file:
    content = file.read()
    print("Before writing:")
    print(content)

    # move the file cursor to the end of the file
    file.seek(0,2)  # 2 means end of the file
    file.write("This line is added in r+ mode.\n")

    # move the file cursor to the beginning to read the updated content
    file.seek(0)
    updated_content = file.read()
    print("After writing:")
    print(updated_content)
    
```
