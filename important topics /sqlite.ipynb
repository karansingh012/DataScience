```python
import sqlite3
connection = sqlite3.connect('test.db')
connection
```

```python
cursor = connection.cursor()
```

```python
cursor.execute('''
CREATE TABLE IF NOT EXISTS employees(
    id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    age INTEGER,
    department TEXT
)
''')
# commit the changes 
connection.commit()
```

```python
cursor.execute('''
  select*from employees
''')
```

```python
cursor.execute("INSERT INTO employees(name,age,department) VALUES('karan',20,'data science')")
cursor.execute("INSERT INTO employees(name,age,department) VALUES('mayank',20,'data science')")
connection.commit()
cursor.execute("SELECT * FROM employees")
print(cursor.fetchall())
```

```python
# Get only the latest 2 rows
cursor.execute("SELECT * FROM employees ORDER BY id DESC LIMIT 2")
rows = cursor.fetchall()

# Print them using loop
for row in rows:
    print(row)
```

```python
# update the data in table 
cursor.execute('''
update employees
set age = 21
where name="mayank"
''')
# changes the commits
connection.commit()
```
