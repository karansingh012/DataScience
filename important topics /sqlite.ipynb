```python
## connect to and sqlite database
import sqlite3
connection = sqlite3.connect('test.db')
connection
```

```python
cursor = connection.cursor()
```

```python
cursor.execute('''
CREATE TABLE IF NOT EXISTS employees(
    id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    age INTEGER,
    department TEXT
)
''')
# commit the changes 
connection.commit()
```

```python
cursor.execute('''
  select*from employees
''')
```

```python
# Delete all existing data
cursor.execute("DELETE FROM employees")
connection.commit()

# Insert data only once
cursor.execute("INSERT INTO employees(name,age,department) VALUES('karan',20,'data science')")
cursor.execute("INSERT INTO employees(name,age,department) VALUES('mayank',21,'data science')")
connection.commit()

# Query and print
cursor.execute("SELECT * FROM employees")
rows = cursor.fetchall()

for row in rows:
    print(row)
```

```python
# working with sales data 
connection = sqlite3.connect('sales_data.db')
cursor = connection.cursor()

# Drop existing table if it exists
cursor.execute("DROP TABLE IF EXISTS sales")
connection.commit()

# Create fresh table with all columns
cursor.execute("""
create table sales(
    id INTEGER PRIMARY KEY,
    date TEXT NOT NULL,
    product TEXT NOT NULL,
    sales TEXT,
    region TEXT
)
""")

## insert data into sales table 
sales_data = [
    ('11-04-2023','product1',100,'lucknow'),
    ('12-04-2023','product2',200,'delhi'),
    ('13-04-2023','product3',300,'ghaziabad'),
    ('14-04-2023','product4',400,'kanpur'),
    ('15-04-2023','product5',500,'meerut')
]

## to print many line 
cursor.executemany('''
insert into sales(date,product,sales,region) 
                   values(?,?,?,?)
''',sales_data)

connection.commit()
```

```python
## query data from sales table
cursor.execute("SELECT*FROM sales")
rows = cursor.fetchall()

## print queried data 
for row in rows:
    print(row)
```
