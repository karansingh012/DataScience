```python
## reading method 1: read entire file content
with open('example.txt','r') as file:
    content = file.read()
    print(content)
```

```python
# reading method 2: read file line by line
with open('example.txt','r') as file:
    for line in file:
        print(line.strip())  # Using strip() to remove leading/trailing whitespace
```

```python
## writing method: write to a file
with open('main.txt','w') as file:
    file.write("Hello, World!\n")
    file.write("This is a test file.\n")
```

```python
## write a file using append mode means data will be added at the end of the file
with open('main.txt','a') as file:
    file.write("This line is appended to the file.\n") 
    
```

```python
# r+ means read and write mode
with open('example.txt','r+') as file:
    content = file.read()
    print("Current Content:")
    print(content)
    
    # Move the cursor to the end of the file before writing
    file.seek(0, 2)  # 2 means from the end of the file
    file.write("\nAdding a new line at the end of the file.")
    
    ## read the updated content 
    file.seek(0)  # Move the cursor back to the beginning of the file
    updated_content = file.read()
    print("Updated Content:")
    print(updated_content)

```

```python
## w+ means read and write mode
## w+ mode: opens a file for both writing and reading. If the file exists, it truncates the file to zero length. If the file does not exist, it creates a new file.
with open('sample.txt','w+') as file:
    file.write("This is a sample text.")
    file.seek(0)
    content = file.read()
    print(content)



"""
humne dekha 
R reading 
w writing 
a append 
r+ reading + writing
w+ writing + reading
a+ append + reading
"""


## how to read any file 
with open("example.txt","r") as file:
    content = file.read()
    print(content)

## how to write any file
with open("newfile.txt","w") as file:
    file.write("This is a new file created using write mode.\n")
    file.write("It contains some sample text.\n")

## how to append any file
with open("newfile.txt","a") as file:
    file.write("This line is appended to the file.\n")
    file.write("Appending another line to the file.\n")

## how to read and write any file
with open("example.txt","r+") as file:
    content = file.read()
    print("Current Content:")
    print(content)
    
    # Move the cursor to the end of the file before writing
    file.seek(0, 2)  # 2 means from the end of the file
    file.write("\nAdding a new line at the end of the file.")
    
    ## read the updated content 
    file.seek(0)  # Move the cursor back to the beginning of the file
    updated_content = file.read()
    print("Updated Content:")
    print(updated_content)

## how to write and read any file
with open("sample.txt","w+") as file:
    file.write("This is a sample text.")
    file.seek(0)
    content = file.read()
    print(content)
```

```python
## using of seek() method 
## The seek() method in Python is used to change the position of the file cursor within a file. This allows you to move to a specific location in the file for reading or writing operations.
with open("example.txt","r") as file:
    # Move the cursor to the 10th byte from the beginning of the file
    file.seek(10)
    content = file.read(20)  # Read 20 bytes from the current cursor position
    print(content)
```

```python
## operations of file handling
# open a file
# read from a file
# write to a file
# append to a file
# close a file

## readlines() method
# The readlines() method in Python is used to read all the lines of a file and return them as a list of strings, where each string represents a line in the file.
with open("example.txt","r") as file:
    lines = file.readlines()
    for line in lines:
        print(line.strip())  # Using strip() to remove leading/trailing whitespace
```
