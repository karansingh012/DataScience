```python
"""Write a Python function to count the frequency of each character in a
given string and return the output in a dictionary. Example:
char_frequency("HELLO") returns {"H":1, "E":1, "L":2, "O":1}"""

def char_frequency(input_string):  ## hello 
    frequency = {}   ##3 dictionary to store frequency of each character
    for char in input_string:  ## means iterate through each character in the string 
        if char in frequency:  ## means if character is already present in dictionary
            frequency[char] += 1  ## means increment the count by 1
        else:  ## means if character is not present in dictionary
            frequency[char] = 1   ## means add character to dictionary with count 1
    return frequency



# # Example usage
# result = char_frequency("HELLO")
# print(result)  # Output: {'H': 1, 'E': 1, 'L': 2, 'O': 1}


char_frequency("Apple")
```

```python
# A website requires the users to input username and password to register. Construct
# a program to check the validity of password input by users.
# Following are the criteria for checking the password:
# 1. At least 1 letter between [a-z]
# 2. At least 1 number between [0-9]
# 3. At least 1 letter between [A-Z]
# 7
# 7
# 7
# 4. At least 1 character from [$#@]
# 5. Minimum length of transaction password: 6
# 6. Maximum length of transaction password: 12
# Your program should accept a sequence of comma separated passwords and will
# check them according to the above criteria. Passwords that match the criteria are
# to be printed, each separated by a comma
# Explore the working of while, and for loop with examples.

def check_passwords(passwords):
    valid_passwords = []

    for password in passwords.split(','):
        password = password.strip()

        if len(password) < 6 or len(password) > 12:
            continue

        has_lower = False
        has_upper = False
        has_digit = False
        has_special = False

        for ch in password:
            if ch.islower():
                has_lower = True
            elif ch.isupper():
                has_upper = True
            elif ch.isdigit():
                has_digit = True
            elif ch in "$#@":
                has_special = True

        if has_lower and has_upper and has_digit and has_special:
            valid_passwords.append(password)

    return ",".join(valid_passwords)
```

```python
## A list is an ordered, mutable collection of elements enclosed in square brackets [].
my_list = [1, 2, 3, 4, 5]
# List Operations:
# 	•	Access elements
# 	•	Insert
# 	•	Update
# 	•	Delete


# List Methods:
# 	•	append()
# 	•	insert()
# 	•	remove()
# 	•	pop()
# 	•	sort()
# 	•	reverse()

list = [5, 2, 9, 1, 5, 6]
list.append(10)  ## add an element to the end of the list
list.insert(2, 15)  ## insert an element at a specific index
list.remove(5)  ## remove the first occurrence of an element
list.pop()  ## remove and return the last element of the list
list.sort()  ## sort the list in ascending order
list.reverse()  ## reverse the order of elements in the list
```

```python
#A tuple is an ordered, immutable collection enclosed in parentheses ().
my_tuple = (1, 2, 3, 4, 5)
# Tuple Operations:
# 	•	Access elements
# 	•	Slicing
# Tuple Methods:
# 	•	count()
# 	•	index()
my_tuple = (1, 2, 3, 4, 5)
my_tuple.count(2)  ## count the occurrences of an element in the tuple
my_tuple.index(3)  ## find the index of the first occurrence of an element in the tuple

## difference between list and tuple
# Lists are mutable, meaning their elements can be changed, added, or removed after creation.
# Tuples are immutable, meaning once they are created, their elements cannot be changed, added
## advantage of tuple over list is that tuples are faster than lists because of immutability
# and they can be used as keys in dictionaries.
# Lists use more memory than tuples due to their dynamic nature.
```

```python
## A dictionary stores data in key–value pairs enclosed in {}.
my_dict = {'name': 'karan', 'age': 25, 'city': 'New York'}
# Dictionary Operations:
# 	•	Access values
# 	•	Insert key-value pairs
# 	•	Update values
# 	•	Delete key-value pairs
# Dictionary Methods:
# 	•	keys()
# 	•	values()
# 	•	items()
# 	•	get()
# 	•	pop()
my_dict = {'name': 'karan', 'age': 25, 'city': 'New York'}
my_dict['age']  ## access the value associated with a key
my_dict['country'] = 'USA'  ## insert a new key-value pair
my_dict['age'] = 26  ## update the value associated with a key
del my_dict['city']  ## delete a key-value pair
my_dict.keys()  ## get a list of all keys in the dictionary 
```
