```python
#np.isnan(array)
import numpy as np 
arr = np.array([1,2,np.nan,4,np.nan,6])
print(np.isnan(arr))
## we can not compear it directilly 
print(np.nan==np.nan)
```

```python
import numpy as np 
arr = np.array([1,2,np.nan,4,np.nan,6])
cleaned_arr = np.nan_to_num(arr,nan=100) ## null value ki jgh humne 100 kr diya hai
print(cleaned_arr)
```

# infinite.py

```python
#np.isinf(array)
arr = np.array([1,2,np.inf,4,-np.inf,6])
print(np.isinf(arr))
cleaned_arr = np.nan_to_num(arr,posinf=1000,neginf=-1000)  
print(cleaned_arr)
```
