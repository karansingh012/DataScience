# Indexing Slicing
# indexing
# array[index] # oneD array
# array [row ,column] # TwoD array

```python
#ek element access krna hai toh indexing agr 1 se jda element access krna hai toh slicing
import numpy as np
arr = np.array([10,20,30,40])
print(arr[0])
print(arr[1])
print(arr[-1])

```

# slicing


```python
import numpy as np
arr = np.array([10,20,30,40,50,60])
print(arr[1:4]) #issme hum negative step bhi pass kr skte hai 
print(arr[:4])
print(arr[::2]) #iska mtlb ye har second element ko pic krega
print(arr[::-1])
```

# fancy indexing

```python
arr = np.array([10,20,30,40,50,60])
print(arr[[0,2,4]]) #it is called fancy indexing
```

# filtiring 

```python
arr = np.array([10,20,30,40,50,60])
print(arr[arr > 25]) # iska mtlb 25 se bde sare element print kra do 
```

# reshaping and mainuplating

```python
# this is use when 1d array convert in 2d and 2d array convert to 3d then we will use reshaping
arr = np.array([10,20,30,40,50,60]) # inhe reshape tbhi kr kste hai jab dimenssions same ho 
reshaped_Arr = arr.reshape(2,3)
print(reshaped_Arr) # reshape does not create copy it is return of view 
```

# flanting array 

```python
"""
.ravel() -> views
.flatten() -> copy
"""
arr_2d = np.array([[12,22,33,44],[22,33,44,55]])
print(arr_2d.ravel)
print(arr_2d.flatten)
```
