# Creating arrays from python list

```python
import numpy as np
arr_1d = np.array([10,20,30,40,50])
print(arr_1d)
```

```python
#with default values 
zeroes_array = np.zeros(3)
print(zeroes_array)
```

```python
#ones(shape)
ones_array = np.ones((2,3))
print(ones_array)
```

```python
#full_(shape,value)
filled_arr = np.full((2,2),7)
print(filled_arr)
```

```python
#creating sequesnce of number in numpy
#arange ()
#arange(start,top,step)
arr = np.arange(1,10,2)
print(arr)
```

```python
#creaating identity matrix
identity_matrix = np.eye(3)
print(identity_matrix)
```

```python
import numpy as np 
## random integer array
arr2 = np.array([12,13,14,15,16,])

def random_integer_array(arr): ## function to calculate mean,median,std
    return np.mean(arr),np.median(arr),np.std(arr)

mean,median,std = random_integer_array(arr2)
print("Mean:",mean)
print("Median:",median)
print("Standard Deviation:",std)
```
