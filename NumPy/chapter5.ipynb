# Broadcasting

```python
# Without adding numpy hum phle dkhte hai ki ye loop se kaise hota hai 
prices = [200,260,270]
discount = 10 #10 perecnt ka discount ho rha hai man lo theek bhai 
final_prices = []
for price in prices:
    final_price = price - (price*discount/100)
    final_prices.append(final_price)
    print(final_prices)

    ## numpy se hum aaise kam without using any loop iss kam ko bhautt easly kr skte hai 
```

```python
## using numpy
import numpy as np
prices = np.array([200,260,270])
discount = 10
final_prices = prices - (prices*discount/100)
print(final_prices)

```

```python
## 1d to 2d array broadcasting
matrix = np.array([[20,30,40],[40,50,60]])
vector = np.array([12,13,14])
result = matrix+vector
print(result) ## dimenssion same hona chaiye theek 
```
