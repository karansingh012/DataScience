# Encapsulation
Encapsulation is the concept of warpping data(vairable) and methods(Functions) together as a single unit. it restricts direct access to some the object's components, which is a mean of preventing accidental interference and misuse of the data

```python
### Encapsulation with getter and setter method 
### public protected private variables or access modifiers 
class Person:
    def __init__(self,name,age):
        self.name=name ## public variable 
        self.age=age ## public variable
def get_name(person):
        return person.name

person = Person("karan",20)
get_name(person)
```
