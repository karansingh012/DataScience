## Inheritance in python

* inheritance in python is a fundamental concept of oop thats allow a class to inheirt attributes and methods rom another class. This chapter overs single inheritance and multiple inheritance, demontrating wo to create and use them in python 

```python
## “Baccha class apne baap class ki sab cheeze automatically le leta hai.”
#parent class
class car:
    def __init__(self,doors,windows,enginetype):
        self.doors=doors
        self.windows=windows
        self.enginetype=enginetype
    def drive(self):
        print(f"the person is drive the {self.enginetype} car")
car1 = car(5,4,"petroltype")
car1.drive()
```

```python
class tata(car):
    def __init__(self,doors,windows,enginetype,Is_selfdriving):
        super().__init__(doors,windows,enginetype)
        self.Is_selfdriving=Is_selfdriving
    def selfdriving(self):
        print(f"tata supporst self driving : {self.Is_selfdriving}")
tata1 = tata(4,5,"petrol",True)
tata1.selfdriving()
```
