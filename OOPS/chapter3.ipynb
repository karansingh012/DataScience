# Polymorphism

Ploymorphism is a core concept in oop hat allow objects of diffrent classes to be treated as ebjects of a common superclass. It provides way to perform a single action in diffrent froms. Polymorphism is typically achived hrough method verriding and nterfaces

# Method Overriding 

Method Overriding allow a child class to provids a single implentation hat is already defined n its parent class

```python
# base class
class Animal:
    def speak(self):
        return "sound of the animal"
# Derived class 1
class Dog(Animal):
    def speak(self):
        return "woof!"
## Derived class 2
class Cat(Animal):
    def speak(self):
        return "meow!"
```

```python
dog = Dog()
cat = Cat()
print(dog.speak())
print(cat.speak())
```

```python
## ploymorphism with function and method 
# base class
class Shape:
    def __init__(self):
        pass

# derived class 1
class Rectangle(Shape):
    def __init__(self,width,height):
        self.width=width
        self.height=height
    def area(self):
        return self.width*self.height

# derived class 2
class Circle(Shape):
    def __init__(self,radius):
        self.radius=radius
    def area(self):
        return 3.14*self.radius*self.radius

## function that demonstrates polymorphism
def print_area(shape):
    print(f"the area is {shape.area()}")
```

```python
rectangle = Rectangle(3,5)
circle = Circle(3)
print_area(rectangle)
print_area(circle)
```

# polymorphism with Abstract base class

Abstract base classes (ABCs) are used to define ommon methods for group of related objects .They an enforce hat derived class mplement perticular methods, promoting consistency across diffrent implemenations

```python
from abc import ABC,abstractmethod

# define an abstarct class 
class Vechile(ABC):
    @abstractmethod
    def start_engine(self):
        pass
## derived class 1
class Car(Vechile):
    def start_engine(self):
        return "CAr engine started"
#3 derived class 2
class Motercycle(Vechile):
    def start_engine(self):
        return "motarcycle engine are started"
    ## function that demonstrates polymorphism
    def start_vechile(Vechile):
        print(Vechile.start_engine)
```
