# Polymorphism

Ploymorphism is a core concept in oop hat allow objects of diffrent classes to be treated as ebjects of a common superclass. It provides way to perform a single action in diffrent froms. Polymorphism is typically achived hrough method verriding and nterfaces

# Method Overriding 

Method Overriding allow a child class to provids a single implentation hat is already defined n its parent class

```python
# base class
class Animal:
    def speak(self):
        return "sound of the animal"
# Derived class 1
class Dog(Animal):
    def speak(self):
        return "woof!"
## Derived class 2
class Cat(Animal):
    def speak(self):
        return "meow!"
```

```python
dog = Dog()
cat = Cat()
print(dog.speak())
print(cat.speak())
```
