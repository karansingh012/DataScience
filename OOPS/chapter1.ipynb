```python
# instance variable and method 
class Dog:
    # constuctor
    def __init__(self,name,age):
        self.name=name
        self.age=age
    # create object
dog1 = Dog("Buddy",3)
print(dog1)
print(dog1.name)
print(dog1.age)
```

```python
dog2 = Dog("Lucy",5)
print(dog2)
print(dog2.name)
print(dog2.age)
```

```python
# define a class with instance method
class Dog:
    def __init__(self,name,age):
        self.name = name 
        self.age = age
    def barks(self):
        print(f'{self.name} say woof')

dog1 = Dog("buddy",3)
dog1.barks()
dog2 = Dog("Mayank",8)
dog2.barks()
dog3 = Dog("KAran",10)
dog3.barks()
```

```python
## modeling a bank account 


# define a class for bak account 
class BankAccount:
    def __init__(self,owner,balance=0):
        self.owner = owner
        self.balance = balance
    def deposit(self,amount):
        self.balance += amount
        print(f"{amount} is deposit new balance is {self.balance}")
    def withdraw(self,amount):
        if amount>self.balance:
            print("Insufficient funds!")
        else:
            self.balance -= amount
            print(f"{amount} is withdraw. new balance is {self.balance}")
    def get_balance(self):
        return self.balance
```

```python
Account1 = BankAccount("Karan Singh",50000)
Account1.get_balance()
## to check bank balance 
```

```python
## call instance method 
Account1.deposit(500)
```

```python
# Withdraw some money
Account1.withdraw(60000)
```
