## Abstraction 
Abstraction is the concept of hiding the complex implementation details and showing only the necessary 
features of an object. this helps in reducing programming complexcity and effort.

```python
from abc import ABC,abstractmethod
class Vechile(ABC):
    def drive(self):
        print("the vechile is use for driving")
    @abstractmethod
    def start_engine(self):
        pass
class Car(Vechile):
    def start_engine(self):
        print("car engine are stared")
def operate_vechile(Vechile):
    Vechile.start_engine()

car = Car()
operate_vechile(car)
```
